```python
import numpy as np
import pandas as  pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

**Data Preprocessing**

```python
raw_mail_data=pd.read_csv('spam.csv',encoding = "ISO-8859-1")
#replace the null value with null string
mail_data=raw_mail_data.where((pd.notnull(raw_mail_data)), '')
```

```python
mail_data.shape
```

```python
#Spam mail as 0 ; Non-spam mail as 1
mail_data.rename(columns = {'v1':'Category'}, inplace = True)

```

```python
mail_data
```

```python
mail_data.head()
```

```python
mail_data.loc[mail_data["Category"]=='spam','Category',]=0
mail_data.loc[mail_data["Category"]=='ham','Category',]=1
```

```python
mail_data
```

```python
mail_data.loc[mail_data['Category']=='spam','Category',]=0
mail_data.loc[mail_data['Category']=='ham','Category',]=1
```

```python
mail_data.head()
```

```python
mail_data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
X=mail_data["v2"]
Y=mail_data["Category"]
```

```python
print(X)
print('**********************************')
print(Y)
```

Train Test Split

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,test_size=0.2,random_state=3)
```

Feature Extraction

```python
#Transform the text to feature vectors (input to SVM Model )
# Convert text to lower case letter
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase='True')
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)
#convert y_train and Y_test as int
Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

Training the model->SVM

```python
model=LinearSVC()
model.fit(X_train_features,Y_train)
```

Evaluation of model

```python
pred_on_training_data=model.predict(X_train_features)
accuracy_on_training_data=accuracy_score(Y_train,pred_on_training_data)
```

```python
print('Accuracy on training data:',pred_on_training_data)
```

**Prediction on new mail**

```python
input_mail=["Ok lar... Joking wif u oni...,,,"]
#Convert text to feature vectors
input_mail_features=feature_extraction.transform(input_mail)
prediction=model.predict(input_mail_features)
```
